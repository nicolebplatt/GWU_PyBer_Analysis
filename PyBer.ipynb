```python
# merge dfs
# For merging data frames: inner merge is default unless otherwise specified
pyber_df = pd.merge(ride_df,city_df,on=["city","city"],how="left")
pyber_df.head()
```

```python
# Create the city type DataFrames
urban_cities_df = pyber_df[pyber_df["type"]=="Urban"]
rural_cities_df = pyber_df[pyber_df["type"]=="Rural"]
suburban_cities_df = pyber_df[pyber_df["type"]=="Suburban"]
```

```python
# Get the number of rides for each city by city type
urban_count = urban_cities_df.groupby(["city"]).count()["ride_id"]
rural_count = rural_cities_df.groupby(["city"]).count()["ride_id"]
suburban_count = suburban_cities_df.groupby(["city"]).count()["ride_id"]
```

```python
# Get average fare for each city by city type
urban_avg_fare = urban_cities_df.groupby(["city"]).mean()["fare"]
rural_avg_fare = rural_cities_df.groupby(["city"]).mean()["fare"]
suburban_avg_fare = suburban_cities_df.groupby(["city"]).mean()["fare"]
```

```python
# Get the average number of drivers for each city by city type
urban_drivers = urban_cities_df.groupby(["city"]).mean()["driver_count"]
rural_drivers = rural_cities_df.groupby(["city"]).mean()["driver_count"]
suburban_drivers = suburban_cities_df.groupby(["city"]).mean()["driver_count"]
```

```python
# Build the scatter plots for urban cities
plt.scatter(urban_count,urban_avg_fare,color='coral',s=[i*10 for i in urban_drivers],edgecolors='black',
           linewidths=1, alpha=.8,label='Urban')
# could have also used s=10*urban_drivers
plt.legend()
plt.grid()
plt.title('PyBer Ride-Sharing Data (2019)')
plt.ylabel('Average Fare ($)')
plt.xlabel('Total Number of Rides (Per City)')
plt.show()
```

```python
# Build the scatter plots for suburban cities
plt.scatter(suburban_count,suburban_avg_fare,c='skyblue',s=10*suburban_drivers,edgecolors='black',
           linewidths=1, alpha=.8,label='Suburban')
plt.legend()
plt.grid()
plt.title('PyBer Ride-Sharing Data (2019)')
plt.ylabel('Average Fare ($)')
plt.xlabel('Total Number of Rides (Per City)')
plt.show()
```

```python
# Build the scatter plots for rural cities
plt.scatter(rural_count,rural_avg_fare,c='gold',s=10*rural_drivers,edgecolors='black',
           linewidths=1, alpha=.8,label='Rural')
plt.legend()
plt.grid()
plt.title('PyBer Ride-Sharing Data (2019)')
plt.ylabel('Average Fare ($)')
plt.xlabel('Total Number of Rides (Per City)')
plt.show()
```

```python
# Combining all three scatterplots
plt.subplots(figsize=(10,6))
plt.scatter(urban_count,urban_avg_fare,c='coral',s=10*urban_drivers,edgecolors='black',
           linewidths=1, alpha=.8,label='Urban')
plt.scatter(suburban_count,suburban_avg_fare,c='skyblue',s=10*suburban_drivers,edgecolors='black',
           linewidths=1, alpha=.8,label='Suburban')
plt.scatter(rural_count,rural_avg_fare,c='gold',s=10*rural_drivers,edgecolors='black',
           linewidths=1, alpha=.8,label='Rural')
plt.legend()
plt.grid()
plt.title('PyBer Ride-Sharing Data (2019)',fontsize=20)
plt.ylabel('Average Fare ($)',fontsize=12)
plt.xlabel('Total Number of Rides (Per City)',fontsize=12)
plt.show()
```
